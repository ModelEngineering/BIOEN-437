This notebook illustrates the workflow for building an application up to the point of constructing a package for distribution.

The application is a tool to help in the evaluation of mechanistic models. The tool automatically generates a sequential pathway of length specified by the user with mass action kinetics and kinetic constants specified by the user. An example of such a pathway is $\$S_1 \xrightarrow{k_1} S_2 \xrightarrow{k_2} S_3 \xrightarrow{k_3} âˆ…$

```python
THREE_SPECIES_PATHWAY = """
$S1 -> S2; k1*S1
S2 -> S3; k2*S2
S3 ->; k3*S3

S1 = 10
S2 = 0
S3 = 0
k1 = 1
k2 = 1
k3 = 1
"""
```

```python
# Evaluate the model in the usual way
rr = te.loada(THREE_SPECIES_PATHWAY)
rr.simulate()
rr.plot()
```
